```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
salary_data = pd.read_csv("Salary.csv")
salary_data
```

# **Matplot Library**

```python
plt.plot(salary_data['Salary'] , color='r')
plt.xlabel('Y_Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.hist(salary_data['Salary'], bins=9, color = 'c')
plt.xlabel('Salary')
plt.ylabel('Y_Experience')
plt.show()
```

```python
plt.xlabel('Salary')
plt.ylabel('YearsExperience')
plt.title('Scatter plot')
plt.scatter(x=salary_data['Salary'],y=salary_data['YearsExperience'], color ='red')
```

```python
sns.histplot(data=salary_data, x='Salary', bins=20, kde=True, color = 'purple')
plt.xlabel('Salary')
plt.ylabel('YearsExperience')
plt.title('Salary Distribution')
plt.show()
```

## **Seaborn Library**

```python
correlation_matrix = salary_data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
salary_comparison = salary_data.groupby('YearsExperience')['Salary'].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(data=salary_data, x='YearsExperience', y='Salary')
plt.xlabel('Experience')
plt.ylabel('Average Salary')
plt.title('Average Salary by Experience')
plt.show()
```

# **SKlearn Library**

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
X = salary_data[['YearsExperience']]
y = salary_data['Salary']
model.fit(X, y)
slope = model.coef_[0]
intercept = model.intercept_
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
new_experience = 5
predicted_salary = model.predict([[new_experience]])
```

```python
print(f"Predicted Salary for {new_experience} years of experience: {predicted_salary[0]:,.2f}")
```

```python
sns.scatterplot(data=salary_data, x='YearsExperience', y='Salary', color = 'Black')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Salary vs. Years of Experience')
plt.legend(['Regression'])
plt.plot(X, model.predict(X), color='red', linewidth=2)
plt.show()
```
