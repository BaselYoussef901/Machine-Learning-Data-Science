```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"
cars_df = pd.read_csv('historical_automobile_sales.csv')
cars_df
```

# **Task 1.1**
Develop a *Line chart* using the functionality of pandas to show how automobile sales fluctuate from year to year.

```python
plt.figure(figsize=(12,6))
plt.plot(cars_df['Year'], cars_df['Price'], marker='o', color='b')
plt.title('Automobile Sales over the Years')
plt.xlabel('Year')
plt.ylabel('Price')


plt.show()
```

# **Task 1.2**
Plot different lines for categories of `Vehicle Type` and analyse the trend to answer the question "Is there a noticeable difference in sales trends between different vehicle types during recession periods?

```python
plt.figure(figsize=(15,6))
for V_Type in cars_df['Vehicle_Type'].unique():
    for is_recession in [True,False]:
        data = cars_df[ (cars_df['Vehicle_Type'] == V_Type) & (cars_df['Recession'] == is_recession) ]
        label = f'{V_Type} (Recession)' if is_recession else f'{V_Type} (Non-Recession)'
        linestyle = '-' if is_recession else '--'
        plt.plot(data['Year'] , data['Price'] , marker='o' , linestyle=linestyle , label=label)

plt.title('Sales Trends by Vehicle Type During Recession Periods')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

# **Task 1.3**
Use the functionality of Seaborn Library to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.

```python
import seaborn as sns
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot(data=cars_df, x='Year', y='Price', hue='Vehicle_Type', style='Recession', markers=True, dashes=False)
plt.title('Sales Trend Comparison (Recession vs. Non-Recession)')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.legend(title='Recession', loc='upper right')
plt.show()
```

# **Task 1.4**
 Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period
 

```python
fig, axs = plt.subplots(2, 1, figsize=(10, 8), sharex=True)
# Plot for recession period
recession_data = cars_df[cars_df['Recession'] == 1]
axs[0].plot(recession_data['Year'], recession_data['GDP'], marker='o', linestyle='-', color='red')
axs[0].set_title('GDP Variation During Recession Period')
axs[0].set_ylabel('GDP')
axs[0].grid(True)

# Plot for non-recession period
non_recession_data = cars_df[cars_df['Recession'] == 0]
axs[1].plot(non_recession_data['Year'], non_recession_data['GDP'], marker='o', linestyle='-', color='green')
axs[1].set_title('GDP Variation During Non-Recession Period')
axs[1].set_xlabel('Year')
axs[1].set_ylabel('GDP')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

# **Task 1.5**
Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales

```python
# Assign colors to each month (optional)
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'pink', 'brown', 'gray', 'cyan', 'magenta', 'black']
#bubble_size = [s*10 for s in cars_df['Price']]


from matplotlib.cm import get_cmap
cmap = get_cmap('tab20c', 528)
normalized_prices = (cars_df['Price'] - cars_df['Price'].min()) / (cars_df['Price'].max() - cars_df['Price'].min())

plt.figure(figsize=(12, 6))
plt.scatter(cars_df.index, cars_df['Month'], s=100, c=cmap(normalized_prices), alpha=0.7, edgecolors='k', linewidths=1)

plt.title('Impact of Seasonality on Automobile Sales')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(cars_df.index, cars_df['Month'] , rotation=90) 

plt.show()
```

# **Task 1.7**
Create a pie chart to display the portion of advertising expenditure of XYZAutomotives during recession and non-recession periods

```python
# Calculate total expenditure during recession and non-recession periods
total_recession_expenditure = cars_df[cars_df['Recession']==1]['Advertising_Expenditure'].sum()
total_non_recession_expenditure = cars_df[cars_df['Recession']==0]['Advertising_Expenditure'].sum()

# Create data for the pie chart
expenditure_data = [total_recession_expenditure, total_non_recession_expenditure]
labels = ['Recession', 'Non-Recession']
colors = ['yellow', 'cyan']
explode = (0.1, 0)  # Explode the first slice (Recession)

# Create the pie chart
plt.figure(figsize=(5, 5))
plt.pie(expenditure_data, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', startangle=140)

plt.title('Advertising Expenditure of XYZAutomotives During Recession and Non-Recession Periods')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

# **Task 2.4**
Creating Callbacks; Define the callback function to update the input container

```python
# Define a callback to update the plot based on user input
@app.callback(
    Output('sales-plot', 'figure'),
    Input('vehicle-type-dropdown', 'value'),
    Input('year-dropdown', 'value')
)
def update_plot(selected_vehicle_type, selected_year):
    # Filter the DataFrame based on the selected vehicle type and year
    filtered_df = cars_df[(cars_df['Vehicle_Type'] == selected_vehicle_type) & (cars_df['Year'] == selected_year)]

    # Create a plot using Plotly Express (you can use your preferred plotting library)
    fig = px.bar(filtered_df, x='Year', y='Sales', title=f'Sales for {selected_vehicle_type} in {selected_year}')
    
    return fig
```

## **Task 2.6**
Create and display graphs for Yearly Report Statistics

**Making our Yearly Data**

```python
# Yearly Total Profit
Yearly_Total_Profit = cars_df.groupby('Year')['Price'].sum().reset_index()

# Yearly Total Sold Cars
Yearly_Total_Sold_Cars = cars_df.groupby('Year')['Automobile_Sales'].sum().reset_index()

# Yearly Total Advertising
Yearly_Total_Advertising = cars_df.groupby('Year')['Advertising_Expenditure'].sum().reset_index()


# Yearly Total Competition
Yearly_Total_Competition = cars_df.groupby('Year')['Competition'].sum().reset_index()

# Yearly Buying City
total_city_buying_year = cars_df.groupby(['Year' , 'City']).size().reset_index(name='Total Buying')
yearly_total_buying_cities = total_city_buying_year.groupby(['Year' , 'City'])['Total Buying'].idxmax()
unique_years_with_highest_buying_city = total_city_buying_year.loc[yearly_total_buying_cities]
years_with_highest_buying_city = unique_years_with_highest_buying_city.sort_values(by=['Year', 'Total Buying'], ascending=[True, False])
Yearly_Buying_City = years_with_highest_buying_city.drop_duplicates(subset='Year').reset_index()
Yearly_Buying_City = Yearly_Buying_City.drop('index' , axis=1)


Yearly_Buying_City
```

```python
app.layout = html.Div([
    html.H1('Yearly Report Statistics', style={'textAlign': 'center'}),
    
    # Dropdown for selecting a statistic
    dcc.Dropdown(
        id='statistic-dropdown',
        options=[
            {'label': 'Yearly Profit', 'value': 'Price'},
            {'label': 'Yearly Sold Cars', 'value': 'Automobile_Sales'},
            {'label': 'Yearly Advertising', 'value': 'Advertising_Expenditure'},
            {'label': 'Yearly Competition', 'value': 'Competition'},
            #{'label': 'Yearly Buying City', 'value': 'Yearly_Buying_City'},
        ],
        value='Price'  # Default selected value
    ),
    
    # Line chart for the selected statistic
    dcc.Graph(id='line-chart'),
    
    # Bar chart for yearly statistics
    dcc.Graph(id='bar-chart')
])
```

```python
# Define a callback to update the line chart based on the selected statistic
@app.callback(
    Output('line-chart', 'figure'),
    Input('statistic-dropdown', 'value')
)
def update_line_chart(selected_statistic):
    Data = Yearly_Total_Profit
    if selected_statistic == 'Price':
        Data = Yearly_Total_Profit

    if selected_statistic == 'Automobile_Sales':
        Data = Yearly_Total_Sold_Cars

    if selected_statistic == 'Advertising_Expenditure':
        Data = Yearly_Total_Advertising

    if selected_statistic == 'Competition':
        Data = Yearly_Total_Competition

    fig = px.line(Data, x='Year', y=selected_statistic, title=f'{selected_statistic} Over Years')
    return fig
```

```python
# Define a callback to update the bar chart for yearly statistics
@app.callback(
    Output('bar-chart', 'figure'),
    Input('statistic-dropdown', 'value')
)
def update_bar_chart(selected_statistic):
    Data = Yearly_Total_Profit
    if selected_statistic == 'Price':
        Data = Yearly_Total_Profit

    if selected_statistic == 'Automobile_Sales':
        Data = Yearly_Total_Sold_Cars

    if selected_statistic == 'Advertising_Expenditure':
        Data = Yearly_Total_Advertising

    if selected_statistic == 'Competition':
        Data = Yearly_Total_Competition

    fig = px.bar(Data, x='Year', y=selected_statistic, title=f'{selected_statistic} Over Years')
    return fig
```

```python
if __name__ == '__main__':
    app.run_server(debug=True)
```
